# US Equity Indices Return Breakdown

## S&P 500

```python
import pandas as pd
import numpy as np

import datetime as dt
from datetime import datetime

import seaborn as sns
sns.set_style('whitegrid')
import matplotlib.pyplot as plt
%matplotlib inline

import matplotlib.pyplot as plt
%matplotlib inline

import PyDSWS as PyDSWS
ds = PyDSWS.Datastream(username='X', password='X')

df = ds.get_data(tickers='S&PCOMP',
                  start="-57Y",end="-1", freq="M")

df.columns= ['price']
df = df.round(2)
df
```

```python
df['return'] = ((df['price']-df['price'].shift(1))/df['price'].shift(1))*100
```

```python
return_group = []
for x in df['return']:
    if x < -20:
        return_group.append('< -20%')
    elif x < -10:
        return_group.append('> -20 < -10%')
    elif x < -5:
        return_group.append('> -10 < -5%')
    elif x < 0:
        return_group.append('> -5% < 0%')
    else:
        return_group.append('> 0%')
        
df['return bucket'] = return_group

df_ = df.groupby(['return bucket']).mean().round()

df_
```

```python
df_ = df.groupby(['return bucket']).count()
df_
```

```python
df_['% of Total Down Month'] = [x for x in df_['return']/259*100]
```

```python
df_.round()
```
